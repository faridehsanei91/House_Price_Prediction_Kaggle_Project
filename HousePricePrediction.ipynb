<span style="font-size: 20px">Encoding Categorical Variables</span>

```python
y = df_train['SalePrice']
X = df_train.drop('SalePrice', axis=1)
```

```python
combined_df = pd.concat([X, df_test], keys=['train', 'test'])
```

```python
combined_df = pd.get_dummies(combined_df, drop_first=True)
```

```python
combined_df
```

```python
X = combined_df.xs('train')
df_test = combined_df.xs('test')
```

```python
print(X.shape)
print(df_test.shape)
```

<span style="font-size: 40px">Model Building and Evaluation</span>

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.20, random_state=101)
```

<span style="font-size: 20px">Data Normalization</span>

<span style="font-size: 15px">This ensures that features are normalized, which is crucial for many machine learning models that are sensitive to the scale of the input features.</span>

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X_train = scaler.fit_transform(X_train)

X_val = scaler.transform(X_val)
```

<span style="font-size: 20px">Machine Learning Model Training</span>

```python
from sklearn.linear_model import Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor


models = {
    'DecisionTree': DecisionTreeRegressor(random_state=42),
    'Ridge': Ridge(),
    'Lasso': Lasso(),
    'RandomForest': RandomForestRegressor(random_state=42),
    'XGBoost': XGBRegressor(random_state=42)
}
```

<span style="font-size: 20px">Model Evaluation</span>

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error


model_performance_rmse = {}
model_performance_mae = {}
model_performance_mse = {}
model_performance_score = {}
model_performance_model= {}

for name, model in models.items():
    
    #trainign model
    model.fit(X_train, y_train)
    model_performance_model[name]= model
    
    
    #evaluating model
    score = model.score(X_val, y_val)
    y_pred = model.predict(X_val)
    mse = mean_squared_error(y_val, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_val, y_pred)
    
    
    model_performance_score[name] = score
    model_performance_mse[name] = mse
    model_performance_rmse[name] = rmse
    model_performance_mae[name] = mae
    
    #printing model performance
    print(f"{name} Performance Metrics:")
    print(f"R Score: {score:.4f}")
    print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
    print(f"Mean Squared Error (MSE): {mse:.4f}")
    print(f"Mean Absolute Error (MAE): {mae:.4f}")
    print("\n")
```

```python
predictions = pd.DataFrame(
    {'DecisionTree': model_performance_model['DecisionTree'].predict(X_val),
     'Ridge': model_performance_model['Ridge'].predict(X_val),
     'Lasso': model_performance_model['Lasso'].predict(X_val),
     'RandomForest': model_performance_model['RandomForest'].predict(X_val),
     'XGBoost': model_performance_model['XGBoost'].predict(X_val)
    })

predictions
```

<span style="font-size: 40px">Model Selection and Final Prediction</span>

```python
best_model_name = max(model_performance_score, key=model_performance_score.get)
best_model = models[best_model_name]
```

```python
best_model
```

```python
xgb_prediction_val = model_performance_model[best_model_name].predict(X_val)
```

```python
plt.scatter(y_val, xgb_prediction_val, label='Predictions')  # Add label to scatter plot
plt.plot(y_val, y_val, color='red', linestyle='-', linewidth=2, label='Perfect Prediction')  # Add label to perfect prediction line
plt.xlabel('Y validation')
plt.ylabel('Predicted Y')
plt.legend()
```
